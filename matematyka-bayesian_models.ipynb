# Budowa modelu

Załóżmy, że podejrzewamy zmianę charakterystyki procesu otrzymywania SMS'ów

Interesuje nas średnia ilość SMS'ów przed zmianą, średnia ilość SMS'ów po zmianie i dzień, w którym zmiana nastąpiła.

Szukamy więc 3 parametrów.

$$ C_i \sim \text{Poisson}(\lambda)  $$

$$
\lambda = 
\begin{cases}
\lambda_1  & \text{if } t \lt \tau \cr
\lambda_2 & \text{if } t \ge \tau
\end{cases}
$$

\begin{align}
&\lambda_1 \sim \text{Exp}( \alpha ) \\\
&\lambda_2 \sim \text{Exp}( \alpha )
\end{align}

Zauważmy, że potrzebujemy również parametru $\alpha$, od którego zależą parametry $\lambda$. Taki parametr nazywamy z ang. hyper-parameter lub parent variable. 

Pierwszy parametr, będzie określał dzień, w którym nastąpiła zmiana średniej liczby SMS'ów.

```python
dzien_zmiany = pm.DiscreteUniform("dzien_zmiany", 1, len(smsData))
```

Kolejne dwie zmienne losowe oznaczać będą średnie ilość SMS'ów

Zauważmy, że aby zdefiniować rozkład Poissona, potrzebujemy parametru $\lambda$.

Zakładamy jednak, że nastąpiła zmiana intensywności procesu, a więc parametr $\lambda$ uległ zmianie.

    lambda_1 = pm.Exponential("lambda_1", alfa)
    lambda_2 = pm.Exponential("lambda_2", alfa)

Jednak możemy dla tego parametru wyznaczyć dobre oszacowanie.
$\alpha$ to odwrotność wartości oczekiwanej, którą wyliczymy z naszych danych.

```python
alfa = 1 / smsData.mean()

lambda_1 = pm.Exponential("lambda_1", alfa)
lambda_2 = pm.Exponential("lambda_2", alfa)

@pm.deterministic
def lambda_final(dzien_zmiany=dzien_zmiany, lambda_1 = lambda_1, lambda_2 = lambda_2):
    lf = np.zeros(len(smsData))
    lf[:dzien_zmiany] = lambda_1
    lf[dzien_zmiany:] = lambda_2
    return lf
```

```python
# Dodajemy obserwacje do zmiennej losowej

zaobserwowane = pm.Poisson("obs", lambda_final, value=smsData, observed=True)
```

```python
# W końcu budujemy model, przekazując definicje naszych zmiennych

model = pm.Model([zaobserwowane, lambda_1, lambda_2, dzien_zmiany])
```
